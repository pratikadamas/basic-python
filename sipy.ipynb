```python

import matplotlib.pyplot as plt

# time axis
t = np.linspace(0, 1, 500)

# signal = two sine waves
signal = np.sin(2*np.pi*2*t) + np.sin(2*np.pi*5*t)

plt.plot(t, signal)
plt.title("Time Domain Signal")
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.show()
```

```python
import numpy as np

# Coefficient matrix
A = np.array([[1, 2, 5],
              [2, 5, 1],
              [2, 3, 8]], dtype=float)

# Constant vector
B = np.array([10, 8, 3], dtype=float)

# Step 1: Find inverse of A
A_inv = np.linalg.inv(A)

print("Inverse of A:")
print(A_inv)

# Step 2: Multiply inverse with B
X = A_inv @ B

print("\nSolution vector X (x, y, z):")
print(X)

# Step 3: Verification (A @ X should equal B)
check = A @ X

print("\nVerification (A @ X):")
print(check)

print("\nOriginal B:")
print(B)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.fft import fft, fftfreq
fs = 100        # Sampling frequency (Hz)
T = 1 / fs
N = 1000        # Number of samples

t = np.arange(N) * T

signal = np.sin(2*np.pi*5*t) + 0.5*np.sin(2*np.pi*12*t)
```

```python
fft_values = fft(signal)
frequencies = fftfreq(N, T)
```

```python
plt.figure()
plt.plot(t, signal)
plt.title("Time Domain Signal")
plt.xlabel("Time (seconds)")
plt.ylabel("Amplitude")
plt.show()
```

```python
plt.figure()
plt.plot(frequencies[:N//2], np.abs(fft_values[:N//2]))
plt.title("Frequency Domain (FFT)")
plt.xlabel("Frequency (Hz)")
plt.ylabel("Magnitude")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

fs = 1000        # Sampling frequency (Hz)
T = 1            # Duration (seconds)
t = np.linspace(0, T, fs, endpoint=False)
f1 = 5    # 5 Hz
f2 = 10   # 10 Hz

signal1 = np.sin(2 * np.pi * f1 * t)
signal2 = np.sin(2 * np.pi * f2 * t)
signal = signal1 + signal2
plt.figure(figsize=(10, 4))
plt.plot(t, signal)
plt.title("Time-Domain Signal (5 Hz + 10 Hz)")
plt.xlabel("Time (seconds)")
plt.ylabel("Amplitude")
plt.xlim(0, 0.5)   # Zoom for clarity
plt.grid()
plt.show()

```

```python
fft_vals = np.fft.fft(signal)
fft_vals = np.abs(fft_vals) / fs   # Normalize

freqs = np.fft.fftfreq(len(fft_vals), 1/fs)
plt.figure(figsize=(10, 4))
plt.stem(freqs[:fs//2], fft_vals[:fs//2])
plt.title("Frequency-Domain Representation")
plt.xlabel("Frequency (Hz)")
plt.ylabel("Magnitude")
plt.xlim(0, 20)
plt.grid()
plt.show()
```

```python
from scipy.fft import  fft,fftfreq
```

```python
# Signal parameters
f1 = 5  # Frequency 1 in Hz
f2 = 10 # Frequency 2 in Hz
duration = 1 # Duration in seconds
sampling_rate = 1000 # Sampling rate in Hz

# Time array
N = int(sampling_rate * duration) # Number of sample points
T = 1.0 / sampling_rate # Sample spacing
t = np.linspace(0.0, duration, N, endpoint=False)
```

```python
# Combined signal: sum of two sine waves
y =  np.sin(2 * np.pi * f1 * t) +  np.sin(2 * np.pi * f2 * t)
```

```python
# Compute FFT
yf = fft(y)
xf = fftfreq(N, T)[:N//2] # Frequency axis for positive frequencies
# Compute magnitude spectrum (absolute value of FFT results, scaled)
magnitude = 2.0/N * np.abs(yf[0:N//2])
```

```python
# Plotting
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(8, 6))

# Time domain plot
ax1.plot(t, y)
ax1.set_xlabel('Time (s)')
ax1.set_ylabel('Amplitude')
ax1.set_title('Time-domain signal (5 Hz + 10 Hz)')
ax1.set_xlim(0, 1)
ax1.grid(True)

# Frequency domain plot
ax2.plot(xf, magnitude)
ax2.set_xlabel('Frequency (Hz)')
ax2.set_ylabel('Magnitude')
ax2.set_title('Frequency-domain representation')
ax2.set_xlim(0, 20) # Limit x-axis to show relevant frequencies
ax2.grid(True)

plt.tight_layout()
plt.show()
```
