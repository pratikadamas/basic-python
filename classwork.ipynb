```python
import exifread

def convert_to_degrees(value):
    d = value.values[0].num / value.values[0].den
    m = value.values[1].num / value.values[1].den
    s = value.values[2].num / value.values[2].den
    return d + (m / 60.0) + (s / 3600.0)

def extract_image_metadata(image_path):
    with open(image_path, 'rb') as f:
        tags = exifread.process_file(f)

    result = {}

    # ---------- CAMERA INFO ----------
    make = tags.get("Image Make")
    model = tags.get("Image Model")
    software = tags.get("Image Software")

    result["camera_make"] = str(make) if make else None
    result["camera_model"] = str(model) if model else None
    result["software"] = str(software) if software else None

    # ---------- OS SEGMENTATION ----------
    if make:
        make_str = str(make).lower()
        if "apple" in make_str:
            result["os"] = "iOS"
        else:
            result["os"] = "Android"
    else:
        result["os"] = "Unknown"

    # ---------- GPS INFO ----------
    lat = tags.get("GPS GPSLatitude")
    lat_ref = tags.get("GPS GPSLatitudeRef")
    lon = tags.get("GPS GPSLongitude")
    lon_ref = tags.get("GPS GPSLongitudeRef")

    if lat and lon and lat_ref and lon_ref:
        latitude = convert_to_degrees(lat)
        if lat_ref.values != "N":
            latitude = -latitude

        longitude = convert_to_degrees(lon)
        if lon_ref.values != "E":
            longitude = -longitude

        result["latitude"] = latitude
        result["longitude"] = longitude
    else:
        result["latitude"] = None
        result["longitude"] = None

    # ---------- COMPRESSION TYPE ----------
    # if "JPEGThumbnail" in tags or image_path.lower().endswith(".jpg"):
    #     result["compression"] = "JPEG"
    # elif image_path.lower().endswith(".png"):
    #     result["compression"] = "PNG"
    # else:
    #     result["compression"] = "Unknown"

    return result


# --------- USAGE ---------
image_file = r"C:\Users\Pratik Giri\OneDrive\Desktop\udb\images\IMG20260305094522.jpeg"
metadata = extract_image_metadata(image_file)

for k, v in metadata.items():
    print(f"{k}: {v}")
```
