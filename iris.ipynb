```python
import matplotlib.pyplot as plt
import numpy as np

# 1. Define the function
def my_function(x):
    return x*x+2*x+3

# 2. Generate x-values
x_values = np.linspace(-20,20,100)

# 3. Calculate y-values
y_values = my_function(x_values)

# 4. Create the plot
plt.figure(figsize=(8, 6)) # Optional: Set figure size
plt.plot(x_values, y_values, label="x*x+2*x+3", color='green', linestyle='--')

# 5. Add labels, title, and legend
plt.title("Plot of a Trigonometric Function")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True) # Optional: Add a grid
plt.legend() # Optional: Display the legend

# 6. Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# 1. Define the function
def my_function(x):
    return 5*np.sin(4*np.pi*x)  

# 2. Generate x-values
x_values = np.linspace(-100,100,100)

# 3. Calculate y-values
y_values = my_function(x_values)

# 4. Create the plot
# plt.figure(figsize=(8, 6)) # Optional: Set figure size
plt.plot(x_values, y_values, label="5*np.sin(4*np.pi*x) ", color='red', linestyle='-')

# 5. Add labels, title, and legend
plt.title("Plot of a Trigonometric Function")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True) # Optional: Add a grid
plt.legend() # Optional: Display the legend

# 6. Show the plot
plt.show()
```
