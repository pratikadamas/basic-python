```python
import numpy as np

def hyperbola_points(a, b, t_range=(-2, 2), orientation='horizontal', num_points=200):
    t = np.linspace(*t_range, num_points)
    if orientation == 'horizontal':
        x = a * np.cosh(t)
        y = b * np.sinh(t)
    else:
        x = a * np.sinh(t)
        y = b * np.cosh(t)
    return x, y
```

```python
import matplotlib.pyplot as plt

a, b = 3, 2
x1, y1 = hyperbola_points(a, b, orientation='horizontal')
x2, y2 = -x1, -y1  # the opposite branch

plt.plot(x1, y1, label='Right branch')
plt.plot(x2, y2, label='Left branch')
plt.axhline(0, color='gray', linestyle='--')
plt.axvline(0, color='gray', linestyle='--')
plt.title('Hyperbola: x²/a² - y²/b² = 1')
plt.legend()
plt.grid(True)
plt.axis('equal')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def compound_function(x):
    # Parabola for |x| < 2
    if abs(x) < 2:
        return x**2 - 1   # simple parabola
    # Hyperbola for |x| >= 2
    else:
        return 4 / x      # simple hyperbola

# Generate values
x_vals = np.linspace(-20, 20, 500)
y_vals = [compound_function(x) for x in x_vals]

# Plot
plt.figure(figsize=(8, 6))
plt.plot(x_vals, y_vals, label='Compound Function', color='royalblue')
plt.axvline(x=0, color='gray', linestyle='--', linewidth=0.8)
plt.axhline(y=0, color='gray', linestyle='--', linewidth=0.8)
plt.title("Compound Function: Parabola + Hyperbola")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def triangle_in_circle(radius=5, center=(0, 0)):
    cx, cy = center

    # Angles for triangle vertices (in radians)
    angles = np.sort(np.random.uniform(0, 2 * np.pi, 3))

    # Points on the circle
    x_points = cx + radius * np.cos(angles)
    y_points = cy + radius * np.sin(angles)

    # Close the triangle by repeating the first vertex
    x_points = np.append(x_points, x_points[0])
    y_points = np.append(y_points, y_points[0])

    # Circle points
    theta = np.linspace(0, 2 * np.pi, 500)
    circle_x = cx + radius * np.cos(theta)
    circle_y = cy + radius * np.sin(theta)

    # Plot
    plt.figure(figsize=(6, 6))
    plt.plot(circle_x, circle_y, 'b', label='Circle')
    plt.plot(x_points, y_points, 'r-', linewidth=2, label='Triangle')
    plt.scatter(x_points[:-1], y_points[:-1], color='red', zorder=5)

    plt.gca().set_aspect('equal', adjustable='box')
    plt.title('Triangle Inside Circle')
    plt.legend()
    plt.grid(True)
    plt.show()

# Example usage
triangle_in_circle()
```

```python
import turtle
import math

def draw_triangle_in_circle(radius):
    # Set up the screen and turtle
    screen = turtle.Screen()
    screen.setup(600, 600)
    t = turtle.Turtle()
    t.speed(5)
    
    # Draw the circle
    t.penup()
    t.goto(0, -radius)  # Start at the bottom of the circle
    t.pendown()
    t.circle(radius)
    
    # Calculate triangle vertices (equilateral triangle)
    angles = [90, 210, 330]  # Angles for vertices in degrees
    vertices = []
    for angle in angles:
        # Convert degrees to radians and calculate position
        x = radius * math.cos(math.radians(angle))
        y = radius * math.sin(math.radians(angle))
        vertices.append((x, y))
    
    # Draw the triangle
    t.penup()
    t.goto(vertices[0])  # Move to first vertex
    t.pendown()
    for vertex in vertices[1:] + [vertices[0]]:  # Include first vertex at end to close triangle
        t.goto(vertex)
    
    # Hide the turtle and display the result
    t.hideturtle()
    screen.exitonclick()

# Example usage
draw_triangle_in_circle(100)
```

```python
import math

def calculate_circumradius(a, b, c):
    """
    Calculates the radius (R) of the circle (circumcircle) 
    that encloses a triangle with side lengths a, b, and c.

    Args:
        a (float): Length of side a.
        b (float): Length of side b.
        c (float): Length of side c.

    Returns:
        float: The circumradius R, or None if the side lengths 
               do not form a valid triangle.
    """
    
    # --- 1. Validation (Check if a valid triangle can be formed) ---
    # The sum of any two sides must be greater than the third side.
    if not (a + b > c and a + c > b and b + c > a):
        print("Error: The provided side lengths do not form a valid triangle.")
        return None

    # --- 2. Calculate the Semi-Perimeter (s) ---
    # s = (a + b + c) / 2
    s = (a + b + c) / 2

    # --- 3. Calculate the Area (K) using Heron's Formula ---
    # K = sqrt(s * (s - a) * (s - b) * (s - c))
    try:
        area_squared = s * (s - a) * (s - b) * (s - c)
        K = math.sqrt(area_squared)
    except ValueError:
        # This catch should theoretically be unnecessary due to the initial validation,
        # but handles cases where rounding errors might cause a slight negative result.
        return None 

    # Handle the case of a degenerate triangle (area is zero)
    if K == 0:
        print("Error: Degenerate triangle (sides are collinear). Cannot determine radius.")
        return None

    # --- 4. Calculate the Circumradius (R) ---
    # R = (a * b * c) / (4 * K)
    R = (a * b * c) / (4 * K)

    return R

# --- Examples ---

# 1. Right Triangle (3-4-5)
# Circumradius of a right triangle is half the hypotenuse (5/2 = 2.5)
r1 = calculate_circumradius(a=3, b=4, c=5)
print(f"Triangle (3, 4, 5): Circumradius R = {r1}") 
# Expected Output: 2.5

# 2. Equilateral Triangle (side=6)
# Area K = 15.588. R should be about 3.464
r2 = calculate_circumradius(a=6, b=6, c=6)
print(f"Triangle (6, 6, 6): Circumradius R = {r2:.3f}") 
# Expected Output: 3.464

# 3. Invalid Triangle
r3 = calculate_circumradius(a=1, b=2, c=10)
print(f"Triangle (1, 2, 10): Circumradius R = {r3}") 
# Expected Output: Error message, returns None
```
